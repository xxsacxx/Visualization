```python
import pandas as pd
from plotnine import *
import numpy as np
```

```python
df=pd.read_csv("pictures.csv")
```

```python
df.head()
```

```python
print(type(df.nominations[0]))
print(type(df.rating[0]))
print(type(df.duration[0]))
```

```python
nomination=df.nominations.astype(int)
```

```python
(ggplot(df, 
        aes('nomination','metacritic')) 
        + geom_point()
        + theme_bw())
```

```python
(ggplot(df, 
            aes('nomination','metacritic', color='factor(genre1)'))
            + geom_point()
            + theme_bw())
```

```python
(ggplot(df, 
        aes('nomination','metacritic', color='factor(genre1)', size='duration'))
        + geom_point() 
        + theme_bw())
```

```python
(ggplot(df, 
        aes('nomination','metacritic', color='factor(genre1)'))
        + geom_point() 
        +  facet_wrap('~release')
        + theme_bw())
```

```python
(ggplot(df, 
        aes('nomination','metacritic', color='factor(genre1)',size='duration'))
        + geom_point() 
        +  facet_wrap('~release')
        + theme_bw())
```
